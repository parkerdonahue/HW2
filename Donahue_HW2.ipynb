```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads\Affairs.csv')
```

```python
print(df.shape[0])
```

```python
print("Number of rows in 'Affairs Dataset': ", df.shape[0])
```

```python
print("Number of columns in 'Affairs Dataset': ", df.shape[1])
```

```python
print(df.types)
```

```python
print(df.dtypes)
```

```python
for col in df.columns:
    if df[col].nunique() == 1:
        print(f"{col} has a constant value")
    elif df[col].nunique() == len(df):
        print(f"unique values for each observation")
```

```python
for col in df.columns:
    mode_values = df[col].mode()
    if col != 'rownames':
        print(f"Mode of {col}:")
        for mode_value in mode_values:
            print(mode_value)
            print("---")
    
```

```python
df_numeric = df.select_dtypes(include=['number'])
print("Mean Values: ", df_numeric.mean())
print("\nMax Values: ", df_numeric.max())
print("\nMin Values: ", df_numeric.min())
```

```python
import matplotlib.pyplot as plt
for col in df_numeric:
    plt.figure(figsize=(10,6))
    df[col].hist(bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    plt.figure(figsize=(10,6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Bar plot of {col}")
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
